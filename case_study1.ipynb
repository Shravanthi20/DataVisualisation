### **6. How do district-level patterns of undernutrition and overnutrition among children vary across India, and what differences emerge from this comparison?**

#### Why This Question Matters  
India faces a **double burden of malnutrition**:  
- **Underweight children** reflect persistent poverty, food insecurity, and poor maternal health.  
- **Overweight children** are emerging due to urbanization, sedentary lifestyles, and nutrition transitions.  

Mapping both patterns side by side reveals **contrasting geographic trends** and highlights regions where **undernutrition and overnutrition overlap** â€” a critical concern for public health planning.  

#### Data Used  
- **Underweight**: Children under 5 years who are underweight (weight-for-age) (%)  
- **Overweight**: Children under 5 years who are overweight (weight-for-height) (%)  
- **Shapefiles**: Contains the districts cordinates of India- dataset owned by the DataMeet GitHub organization.

Both indicators are expressed in percentages, enabling district-level comparisons across India.  

#### Visualization  
Two **choropleth maps** were created using district shapefiles:  
- **Blue map** â†’ Underweight prevalence  
- **Red map** â†’ Overweight prevalence  

This visualization makes the **geographic divergence** between undernutrition and overnutrition.  


```python
# --- Load India district shapefile ---
districts = gpd.read_file(r"Census_2011\2011_Dist.shp")

# Rename long columns for convenience
df.rename(columns={
    'Children under 5 years who are underweight (weight-for-age)18 (%)': 'Underweight',
    'Children under 5 years who are overweight (weight-for-height)20 (%)': 'Overweight'
}, inplace=True)

# --- Clean district names: strip spaces and uppercase ---
districts['DISTRICT_clean'] = districts['DISTRICT'].str.strip().str.upper()
df['District_clean'] = df['District Names'].str.strip().str.upper()


# --- Merge using cleaned names ---
merged = districts.merge(df, left_on='DISTRICT_clean', right_on='District_clean', how='left')


fig, axes = plt.subplots(1, 2, figsize=(18, 10))

merged.plot(
    column='Underweight',
    cmap='Blues',
    linewidth=0.5,
    edgecolor='grey',
    legend=True,
    k=5,                
    ax=axes[0]
)
axes[0].set_title("Underweight Children (%)")
axes[0].axis('off')

merged.plot(
    column='Overweight',
    cmap='Reds',
    linewidth=0.5,
    edgecolor='grey',
    legend=True,
    k=5,
    ax=axes[1]
)
axes[1].set_title("Overweight Children (%)")
axes[1].axis('off')

plt.tight_layout()
plt.show()
```

### Insights
By comparing the two maps side by side:  
- States in the **North and North-East** show higher overweight prevalence.  
- States in the **Central India and Eastern India(West Bengal and Jharkhand)** show higher underweight prevalence.
- This shows India is dealing with two very different nutrition problems simultaneously, depending on region.


### Key takeaway

Undernutrition is a structural, widespread challenge, especially in poorer states.
Overnutrition is a growing, localized issue in more urbanized or unique dietary regions.
Policy interventions need to be region-specific: tackle food insecurity and stunting in Central/Eastern India while addressing obesity and lifestyle risks in the Northeast and urban areas

## Question 7: Which states show the widest gender gap in tobacco or alcohol consumption?

### Why this question matters  
Substance use behaviors such as **tobacco and alcohol consumption** have profound implications for public health. Examining **gender differences** in these habits helps us understand:  
- How **social norms** and **cultural practices** shape menâ€™s and womenâ€™s behaviors differently.  
- Whether **health interventions** need to be gender-targeted.  
- How certain states or regions might face **higher gendered health risks**, creating inequalities in disease burden (e.g., lung disease, cancers, liver problems).  

Understanding the **gender gap** provides insights into **inequalities in exposure to risk factors**, which are often overlooked in policy discussions.  

### Data used  
From the **NFHS-5 District Fact Sheets (2019â€“21)**, aggregated at the state level:  
- **Women age 15+ who use any kind of tobacco (%)**  
- **Men age 15+ who use any kind of tobacco (%)**  
- **Women age 15+ who consume alcohol (%)**  
- **Men age 15+ who consume alcohol (%)**  

Derived variables:  
- `Gap_Tobacco = Men_Tobacco â€“ Women_Tobacco`  
- `Gap_Alcohol = Men_Alcohol â€“ Women_Alcohol`  

These capture the **magnitude of difference** between men and women in each state.  

### Visualization approach  
To highlight gender gaps clearly, we used a **Dumbbell Plot** for both tobacco and alcohol consumption:  
- **Blue dots** represent womenâ€™s percentages.  
- **Red dots** represent menâ€™s percentages.  
- A **connecting line** (dumbbell) shows the **gap** in each state.  
- States are sorted by the widest gap.  
- Color intensity of the line reflects the size of the gender gap.  

This format allows for quick comparison across states while keeping both male and female values visible.  
This analysis not only reveals where **gender disparities in substance use are most pronounced**, but also provides a foundation for **region-specific and gender-sensitive health interventions**.


```python
# --- Data processing steps ---
state_df = df.drop_duplicates(subset=["State/UT"])
state_df = state_df.rename(columns={
    "Women age 15 years and above who use any kind of tobacco (%)": "Women_Tobacco",
    "Men age 15 years and above who use any kind of tobacco (%)": "Men_Tobacco",
    "Women age 15 years and above who consume alcohol (%)": "Women_Alcohol",
    "Men age 15 years and above who consume alcohol (%)": "Men_Alcohol",
    "State/UT": "State"
})
state_df["Gap_Tobacco"] = state_df["Men_Tobacco"] - state_df["Women_Tobacco"]
state_df["Gap_Alcohol"] = state_df["Men_Alcohol"] - state_df["Women_Alcohol"]

# --- Dumbbell Plot for Tobacco Consumption ---
plot_df = state_df.sort_values("Gap_Tobacco", ascending=False).reset_index(drop=True)
fig, ax = plt.subplots(figsize=(10, 12))
norm = plt.Normalize(plot_df["Gap_Tobacco"].min(), plot_df["Gap_Tobacco"].max())
colors = plt.cm.coolwarm(norm(plot_df["Gap_Tobacco"]))
for i, row in plot_df.iterrows():
    ax.plot([row["Women_Tobacco"], row["Men_Tobacco"]],
            [row["State"], row["State"]],
            color=colors[i], linewidth=3, alpha=0.7)
ax.scatter(plot_df["Women_Tobacco"], plot_df["State"],
           color="blue", s=100, label="Women", alpha=0.8)
ax.scatter(plot_df["Men_Tobacco"], plot_df["State"],
           color="red", s=100, label="Men", alpha=0.8)
ax.set_title("Tobacco Use by Gender (%) â€“ NFHS-5 (Dumbbell Plot)", fontsize=14, weight="bold")
ax.set_xlabel("Percentage")
ax.legend()
sns.despine()
plt.tight_layout()
plt.show()

# --- Dumbbell Plot for Alcohol Consumption ---
plot_df = state_df.sort_values("Gap_Alcohol", ascending=False).reset_index(drop=True)
fig, ax = plt.subplots(figsize=(10, 12))
norm = plt.Normalize(plot_df["Gap_Alcohol"].min(), plot_df["Gap_Alcohol"].max())
colors = plt.cm.coolwarm(norm(plot_df["Gap_Alcohol"]))
for i, row in plot_df.iterrows():
    ax.plot([row["Women_Alcohol"], row["Men_Alcohol"]],
            [row["State"], row["State"]],
            color=colors[i], linewidth=3, alpha=0.7)
ax.scatter(plot_df["Women_Alcohol"], plot_df["State"],
           color="blue", s=100, label="Women", alpha=0.8)
ax.scatter(plot_df["Men_Alcohol"], plot_df["State"],
           color="red", s=100, label="Men", alpha=0.8)
ax.set_title("Alcohol Consumption by Gender (%) â€“ NFHS-5 (Dumbbell Plot)", fontsize=14, weight="bold")
ax.set_xlabel("Percentage")
ax.legend()
sns.despine()
plt.tight_layout()
plt.show()
```

### Tobacco Use (Gender Gap)

- Widest Gaps â€“ Nagaland, J&K, Bihar, Jharkhand, Meghalaya (men >50%, women <10%).
- Moderate Gaps â€“ MP, Himachal, Gujarat, Assam, Odisha.
- Narrow Gaps â€“ Southern states & UTs (Kerala, Puducherry, Lakshadweep, Chandigarh).

### Alcohol Use (Gender Gap)

- Widest Gaps â€“ Himachal, Mizoram, Chhattisgarh, Andaman & Nicobar, Arunachal (men 40â€“60%, women <10%).
- Moderate Gaps â€“ Telangana, Jharkhand, Odisha, Nagaland (men 25â€“35%).
- Narrow Gaps â€“ Gujarat, Bihar, J&K, Lakshadweep (low for both genders).

### Combined Takeaways

- Regional Hotspots â€“ Tobacco: NE + Central; Alcohol: NE + Hill States.
- Cultural Influence â€“ Southern & coastal states show consistently lower gender gaps in both behaviors.
- Policy Priority â€“
    - Tobacco: Male-focused interventions in East/Central belt.
    - Alcohol: Hill & tribal states need targeted reduction strategies.

## 9. Do districts that ensure childhood vaccination also provide timely maternal postnatal care, or are these interventions unevenly delivered?

### Why this Question Matters
- Childhood vaccination and maternal postnatal care (PNC) are both **essential interventions** that reduce child and maternal mortality.  
- Examining whether districts that perform well in vaccination also ensure timely PNC reveals whether healthcare delivery is **integrated or fragmented**.  
- Identifying **lagging districts** helps policymakers target areas where both interventions are weak, ensuring more equitable health outcomes.


### Data Used
- **Full Childhood Vaccination Coverage (%)** â€“ Share of children fully immunized.  
- **Maternal Postnatal Care within 2 days (%)** â€“ Share of mothers receiving timely postnatal care after delivery.  
- **District-level data (NFHS-5)**.  

### Visualization
- **Quadrant Scatter Plot** with medians as cut-offs:  
  -  **Good Performers**: High vaccination & high PNC.  
  -  **Needs PNC**: High vaccination but low maternal PNC.  
  -  **Needs Vaccination**: High maternal PNC but low vaccination.  
  -  **Needs Both**: Low vaccination & low PNC.  
- Each district is categorized, and summary counts are shown.  
- The bottom 10 **worst-performing districts** (low vaccination + low PNC) are also highlighted.


```python
# --- Compute medians ---
x_median = plot_df['Full_Vaccination'].median()
y_median = plot_df['Maternal_PNC'].median()

plt.figure(figsize=(12,12))

# --- Quadrant background shading ---
plt.axvspan(0, x_median, 0, y_median/100, facecolor='lightcoral', alpha=0.2)     # Bottom-left (needs both)
plt.axvspan(x_median, 100, 0, y_median/100, facecolor='orange', alpha=0.2)      # Bottom-right (needs PNC)
plt.axvspan(0, x_median, y_median/100, 1, facecolor='gold', alpha=0.2)          # Top-left (needs vaccination)
plt.axvspan(x_median, 100, y_median/100, 1, facecolor='lightgreen', alpha=0.2)  # Top-right (good)

# --- Scatter points colored by quadrant ---
categories = []
colors = []
for i, row in plot_df.iterrows():
    if row['Full_Vaccination'] >= x_median and row['Maternal_PNC'] >= y_median:
        colors.append("green"); categories.append("Good Performers")
    elif row['Full_Vaccination'] >= x_median and row['Maternal_PNC'] < y_median:
        colors.append("orange"); categories.append("Needs PNC")
    elif row['Full_Vaccination'] < x_median and row['Maternal_PNC'] >= y_median:
        colors.append("gold"); categories.append("Needs Vaccination")
    else:
        colors.append("red"); categories.append("Needs Both")

plot_df['Category'] = categories
plt.scatter(plot_df['Full_Vaccination'], plot_df['Maternal_PNC'], 
            c=colors, alpha=0.7, edgecolor="k")

# --- Reference lines ---
plt.axvline(x=x_median, color='red', linestyle='--')
plt.axhline(y=y_median, color='blue', linestyle='--')

# --- Labels and title ---
plt.title("Childhood Vaccination vs Maternal PNC Coverage (NFHS-5)\nQuadrant Analysis of Districts", fontsize=15)
plt.xlabel("Full Childhood Vaccination Coverage (%)", fontsize=13)
plt.ylabel("Maternal Postnatal Care within 2 days (%)", fontsize=13)

# --- Quadrant text annotations ---
plt.text(x_median+1, y_median+18, "Good Performers", color="green", fontsize=12)
plt.text(20, 100, "Needs Vaccination", color="goldenrod", fontsize=12)
plt.text(x_median+1, 25, "Needs PNC", color="darkorange", fontsize=12)
plt.text(10, 25, "Needs Both Interventions", color="darkred", fontsize=12)

plt.grid(alpha=0.3)

# --- Numeric Summary Box ---
summary_counts = plot_df['Category'].value_counts()
summary_text = "\n".join([f"{cat}: {count} districts" for cat, count in summary_counts.items()])

plt.gcf().text(0.02, 0.02, summary_text, fontsize=11, va="bottom", ha="left",
               bbox=dict(facecolor="white", alpha=0.7, edgecolor="gray"))

plt.show()

# --- Print Bottom-left Quadrant (Worst Performers) ---
worst_districts = plot_df[plot_df['Category'] == "Needs Both"].sort_values(
    by=["Full_Vaccination","Maternal_PNC"]
).head(10)

print(" Bottom 10 Worst-performing Districts (Low Vaccination & Low PNC):")
print(worst_districts[['District Names', 'Full_Vaccination', 'Maternal_PNC']])
```

## 10. Where do child malnutrition and maternal health risks collide, and how do education and early marriage shape these high-burden districts?

###  Why this question matters
- **Child and maternal health linkage**: Districts with high stunting or underweight children often coincide with women facing elevated risks of non-communicable diseases (NCDs) like high blood pressure and high blood sugar.  
- **Social determinants**: Female education and early marriage can influence both child malnutrition and maternal health outcomes. Understanding these patterns helps identify districts needing urgent interventions.  
- **Policy relevance**: Highlighting high-burden districts enables targeted health programs and education campaigns to reduce intergenerational health risks.

### Data used
- **Child Malnutrition Indicators**:  
  - Children under 5 years who are stunted (% height-for-age)  
  - Children under 5 years who are underweight (% weight-for-age)  
- **Maternal Health Indicators**:  
  - Women age 15+ with elevated blood pressure (%)  
  - Women age 15+ with high or very high blood sugar (%)  
- **Social Factors**:  
  - Women (15â€“49 years) with â‰¥10 years of schooling (%)  
  - Women (20â€“24 years) married before age 18 (%)  

**Composite indicators created**:  
- `Child_Malnutrition` = Average of stunting & underweight  
- `Maternal_NCD` = Average of high blood pressure & high blood sugar  

**High-burden districts**: Districts above the median for both child malnutrition and maternal NCD risk.

### Visualization
We used **scatter plots and correlation heatmaps** to explore the associations:  
- **Scatter plot**: Child Malnutrition vs. Maternal NCD Risk  
  - Quadrants indicate districts above/below median thresholds.  
  - Red = High risk in both indicators, Gray = Others.  
- **Correlation heatmap**: Shows relationships of child/maternal risks with female education and early marriage.

These visualizations help identify **districts where high child malnutrition and maternal health risks overlap** and highlight the influence of social factors.


```python
# --- Select relevant columns ---
analysis_df = df[['District Names',
                  'Children under 5 years who are stunted (height-for-age)18 (%)',
                  'Children under 5 years who are underweight (weight-for-age)18 (%)',
                  'Women age 15 years and above wih Elevated blood pressure (Systolic â‰¥140 mm of Hg and/or Diastolic â‰¥90 mm of Hg) or taking medicine to control blood pressure (%)',
                  'Women age 15 years and above wih high or very high (>140 mg/dl) Blood sugar level or taking medicine to control blood sugar level23 (%)',
                  'Women (age 15-49)  with 10 or more years of schooling (%)',
                  'Women age 20-24 years married before age 18 years (%)'
                 ]].copy()

# --- Rename for simplicity ---
analysis_df = analysis_df.rename(columns={
    'Children under 5 years who are stunted (height-for-age)18 (%)': 'Stunting',
    'Children under 5 years who are underweight (weight-for-age)18 (%)': 'Underweight',
    'Women age 15 years and above wih Elevated blood pressure (Systolic â‰¥140 mm of Hg and/or Diastolic â‰¥90 mm of Hg) or taking medicine to control blood pressure (%)': 'High_BP',
    'Women age 15 years and above wih high or very high (>140 mg/dl) Blood sugar level or taking medicine to control blood sugar level23 (%)': 'High_BloodSugar',
    'Women (age 15-49)  with 10 or more years of schooling (%)': 'Female_Education10yrs',
    'Women age 20-24 years married before age 18 years (%)': 'Early_Marriage'
})

# --- Clean data: convert to numeric ---
for col in ['Stunting','Underweight','High_BP','High_BloodSugar','Female_Education10yrs','Early_Marriage']:
    analysis_df[col] = pd.to_numeric(analysis_df[col].astype(str).str.replace(r"[^\d\.\-]", "", regex=True), errors="coerce")

# --- Create composite indicators ---
analysis_df['Child_Malnutrition'] = (analysis_df['Stunting'] + analysis_df['Underweight']) / 2
analysis_df['Maternal_NCD'] = (analysis_df['High_BP'] + analysis_df['High_BloodSugar']) / 2

# --- Identify high burden districts ---
malnutrition_thresh = analysis_df['Child_Malnutrition'].median()
ncd_thresh = analysis_df['Maternal_NCD'].median()

analysis_df['Risk_Category'] = analysis_df.apply(
    lambda r: "High Both" if r['Child_Malnutrition'] >= malnutrition_thresh and r['Maternal_NCD'] >= ncd_thresh else "Others",
    axis=1
)

# --- Plot scatter with quadrant thresholds ---
plt.figure(figsize=(10,8))
sns.scatterplot(data=analysis_df, x='Child_Malnutrition', y='Maternal_NCD',
                hue='Risk_Category', palette={'High Both':'red','Others':'gray'}, alpha=0.7)

plt.axvline(malnutrition_thresh, color='red', linestyle='--', label=f"Median Malnutrition = {malnutrition_thresh:.1f}%")
plt.axhline(ncd_thresh, color='blue', linestyle='--', label=f"Median Maternal NCD = {ncd_thresh:.1f}%")

plt.title("Districts with Child Malnutrition vs Maternal NCD Risk (NFHS-5)", fontsize=14)
plt.xlabel("Child Malnutrition (avg of Stunting & Underweight) (%)", fontsize=12)
plt.ylabel("Maternal NCD Risk (avg of High BP & High Blood Sugar) (%)", fontsize=12)
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# --- Correlation heatmap with social factors ---
corr_vars = ['Child_Malnutrition','Maternal_NCD','Female_Education10yrs','Early_Marriage']
plt.figure(figsize=(8,6))
sns.heatmap(analysis_df[corr_vars].corr(), annot=True, cmap="RdYlBu", center=0)
plt.title("Correlation of Child/Maternal Risks with Social Factors", fontsize=13)
plt.show()

# --- Print top high-risk districts ---
high_risk = analysis_df[analysis_df['Risk_Category']=="High Both"].sort_values(
    by=['Child_Malnutrition','Maternal_NCD'], ascending=False
).head(10)

print("Top 10 Districts with High Malnutrition + High Maternal NCD Risk:")
print(high_risk[['District Names','Child_Malnutrition','Maternal_NCD','Female_Education10yrs']])
```

### Key Insights

## Conclusion

The NFHS-5 district-level analysis highlights the deep-rooted **health and social inequalities** across India.  
- Districts with **low female literacy** and **early marriage** consistently report poorer maternal and child health outcomes.  
- **Nutritional challenges** show a dual burden: undernutrition in economically weaker regions and rising overweight/obesity in more urbanized states.  
- Access to **contraception and healthcare services** remains uneven, reflecting systemic gaps in awareness and availability.  

Overall, the data underscores that **health outcomes are inseparable from social determinants**. Addressing inequality requires **integrated interventions** that combine education, healthcare access, and targeted nutrition policies.  
This case study demonstrates how NFHS-5 indicators can guide policymakers toward **district-specific strategies** to reduce disparities and improve population health.
