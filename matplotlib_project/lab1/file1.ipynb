```python
import matplotlib.pyplot as plt
```

```python
x= [1,2,3,4,5]
y= [2,3,3,4,7]
plt.plot(x,y,marker='o',color='black',linestyle='--',linewidth=2,markersize=10,markerfacecolor='r')
plt.title('Plotting line')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
import matplotlib.pyplot as plt
x1=[1,2,3,4,5]
y1=[2,4,6,8,10]
y2=[3,6,9,12,15]
y3= list(map(lambda x:x**2,x1))
plt.plot(x1,y1,marker='o',markerfacecolor='r',linestyle='--',color='purple')
plt.plot(x1,y2,marker='o',markerfacecolor='r',linestyle='--',color='black')
plt.plot(x1,y3,marker='o',markerfacecolor='b',linestyle='--')
plt.title('Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend(["Data Point1", "Data Point2","Data Point3"])
plt.savefig('sample_plot.png')
plt.show()
```
