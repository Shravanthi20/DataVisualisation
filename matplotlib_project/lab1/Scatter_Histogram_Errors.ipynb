```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np

# Load iris dataset into DataFrame
iris_data = load_iris(as_frame=True)
iris = iris_data.frame
iris['species'] = iris_data.target_names[iris['target']]

plt.figure(figsize=(12,5))

# --- Scatter plot (raw data)
plt.subplot(1,2,1)
for species, group in iris.groupby('species'):
    plt.scatter(group['sepal length (cm)'], group['sepal width (cm)'], label=species, alpha=0.7)
plt.title("Scatter Plot: Sepal Length vs Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.legend()

# --- Error bar plot (mean  std)
plt.subplot(1,2,2)
grouped = iris.groupby("species")["sepal width (cm)"]
means = grouped.mean()
stds = grouped.std()

plt.errorbar(x=means.index, y=means.values, yerr=stds.values,
             fmt='o', color='blue', ecolor='red', capsize=5, markersize=8, linewidth=2)
plt.title("Error Bar Plot: Mean Sepal Width  SD")
plt.ylabel("Sepal Width (cm)")
plt.xlabel("Species")

plt.tight_layout()
plt.show()
```

```python
url = 'https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv'
iris = pd.read_csv(url)
iris.head()
```

## 4. Scatter Plot
Let's plot sepal length vs. sepal width.


```python
plt.scatter(iris['sepal_length'], iris['sepal_width'], color='blue', alpha=0.6)
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

## 5. Adding Error Bars
Suppose we have some measurement error. Let's add error bars to the sepal length.


```python
x = iris['sepal_length'][:20]
y = iris['sepal_width'][:20]
yerr = np.random.uniform(0.05, 0.2, size=20)
plt.errorbar(x, y, yerr=yerr, fmt='^', color='green', ecolor='red', capsize=2)#capsize is for length of bar
plt.title('Sepal Width with Error Bars')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

```python
x=iris['petal length (cm)'][:20]
y=iris['sepal length (cm)'][:20]
err= np.random.uniform(0.05,0.2,size=len(y))
plt.errorbar(x, y, yerr=err, fmt='o', color='green', ecolor='red', capsize=2)
```

## 6. Histogram
Let's visualize the distribution of petal length.


```python
plt.hist(iris['sepal_width'], bins=5, color='purple', edgecolor='black', alpha=0.8)
plt.title('Distribution of Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```
