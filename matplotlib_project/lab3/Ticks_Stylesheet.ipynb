# Ticks, Stylesheets: A Data Visualization Tutorial

## 1. Importing Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

## 2. Creating Sample Data


```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

## 3. Plotting Data with Custom Ticks and Stylesheets

```python
plt.figure(figsize=(10, 5))
plt.plot(x, y1, label='Sine Wave', color='blue')
plt.plot(x, y2, label='Cosine Wave', color='orange')
plt.title('Sine and Cosine Waves')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xticks(np.arange(0, 10, 2)) 
plt.yticks(np.arange(-1, 1.5, 0.5)) 
plt.grid(True)
plt.legend()
plt.xlim(0, 10)
plt.ylim(-1, 1)
plt.show()
```

## 4. Applying Stylesheets

```python
plt.style.use('ggplot')  # Applying ggplot style
plt.figure(figsize=(10, 7))
plt.plot(x, y1, label='Sine Wave', color='blue')
plt.plot(x, y2, label='Cosine Wave', color='orange')
plt.title('Sine and Cosine Waves with ggplot Style')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xticks(np.arange(0, 11, 1))  
plt.yticks(np.arange(-1, 2, 0.5))  
plt.legend()
plt.grid(False)
plt.xlim(0, 10)
plt.ylim(-1, 1)
plt.show()
```

## 5. Adding Annotations

```python
plt.figure(figsize=(13, 5))
plt.plot(x, y1, label='Sine Wave', color='blue')
plt.plot(x, y2, label='Cosine Wave', color='orange')
plt.title('Sine and Cosine Waves with Annotations')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xticks(np.arange(0, 11, 1))  # Custom x-ticks
plt.yticks(np.arange(-1, 1.5, 0.5))  # Custom y-ticks
plt.grid(True)
plt.xlim(0, 10)
plt.ylim(-1.2, 1.2)
plt.legend()
plt.annotate('Peak of Sine Wave', xy=(np.pi/2, 1), xytext=(np.pi/2 + 1, 1.2), #xy= coordinates, xytext= where to place text, arrowprops= arrow, shrink= how much the arrow meets both ends
             arrowprops=dict(facecolor='black', shrink=0))
plt.annotate('Peak of Cosine Wave', xy=(0, 1), xytext=(1, 1.2),
             arrowprops=dict(facecolor='black', shrink=0))
plt.show()
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [150, 200, 180, 220, 260, 300]
x_index= sales.index(max(sales))

plt.figure(figsize=(5,5))
plt.plot(months,sales,label='Sales')
plt.title("Monthly Sales Performance")
plt.xlabel('Months')
plt.ylabel('Sales(in units)')
plt.grid(True)
plt.legend("Product A")
plt.annotate("Highest Sales Performance",xy=(x_index,max(sales)),xytext=(x_index+1,max(sales)-1),arrowprops=dict(facecolor='b',shrink=0.5))
```

Requirements:
Plot a line graph of sales vs. months.

Use plt.xticks() to label the x-axis with the month names.

Add a title: "Monthly Sales Performance".

Label the x-axis as "Month" and y-axis as "Sales (in units)".

Add a grid and legend (label="Product A").

Annotate the highest sales month with a label "Highest Sales" and an arrow pointing to it.

Use shrink=0.05 in arrowprops.
